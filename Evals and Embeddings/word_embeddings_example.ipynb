```
# https://radimrehurek.com/gensim/models/word2vec.html
from gensim.models import Word2Vec

yelp_w2v = Word2Vec(yelp['tokenized'], size=150, window=5, min_count=3, workers=2, sg=1, iter=100)
```

```
# There is no way right now to get the average vector per document, but we can create our own function to do that
def get_mean_vector(w2v, words):
    # Make sure words are in the vocabulary---remember, we had dropped all words that occurred less than 3 times in the corpus
    words = [word for word in words if word in w2v.wv.vocab]
    if len(words) >= 1:
        return np.mean(w2v.wv[words], axis=0)
    else:
        return "NC" # I usually use this to indicate no characters
```

```
averaged_doc_vecs = []
y = []
for i, review in enumerate(yelp['tokenized']):
    vec = get_mean_vector(yelp_w2v, review)
    if vec != "NC":
        averaged_doc_vecs.append(vec)
        y.append(yelp.label[i])
```

```
X = np.stack(averaged_doc_vecs, axis=0) # this stacks all the arrays together into one big array that we can use for training
```

```
import random
seed_val = 48109
random.seed(seed_val)
np.random.seed(seed_val)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, shuffle=True)
```

### Logistic Regression

Logistic regression performed the best on the bag-of-words approach, achieving an accuracy of 0.79.

```
from sklearn.metrics import classification_report

# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html
from sklearn.linear_model import LogisticRegression

# Training the model. Notice the regularization is imposed by default. This is different from R.
logreg = LogisticRegression()
clf_logreg = logreg.fit(X_train, y_train)

y_preds_clf_logreg = clf_logreg.predict(X_test)

print(classification_report(y_test, y_preds_clf_logreg))
```

### Gaussian Naive Bayes

In the lab, we used multinomial Naive Bayes. That achieved an accuracy of 0.79. Because we a continuous dense feature space, we use Gaussian Naive Bayes instead.

```
# https://scikit-learn.org/stable/modules/naive_bayes.html
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
clf_nb = nb.fit(X_train, y_train)

y_preds_clf_nb = clf_nb.predict(X_test)

print(classification_report(y_test, y_preds_clf_nb))
```

### Linear Discriminant Analysis

In the bag-of-words example, LDA performed quite terribly. It had an accuracy of 0.61. That's because LDA usually works terribly with sparse matrices. But with word vectors, we now have dense vectors.

```
# https://scikit-learn.org/stable/modules/generated/sklearn.discriminant_analysis.LinearDiscriminantAnalysis.html
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()
clf_lda = lda.fit(X_train, y_train)

y_preds_clf_lda = clf_lda.predict(X_test)

print(classification_report(y_test, y_preds_clf_lda))
```

### Support Vector Machines

SVM had an accuracy of 0.74 on the bag-of-words example. Let's see how SVM does with the word embeddings.

```
# https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html
from sklearn.svm import SVC

sv = SVC()
clf_sv = sv.fit(X_train, y_train)

y_preds_clf_sv = clf_sv.predict(X_test)

print(classification_report(y_test, y_preds_clf_sv))
```

I bet we can do better---let's use cross-validation.

```
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html
from sklearn.model_selection import GridSearchCV

tuning_params = {'C': [1, 10, 100, 1000], 'gamma': [0.001, 0.01, 0.1, 1, 10]}

sv_gridsearch = GridSearchCV(SVC(), tuning_params)
sv_gridsearch_fit = sv_gridsearch.fit(X_train, y_train)
print(sv_gridsearch_fit.best_params_) # this will tell us the best params
```

```
y_preds_clf_sv_gridsearch = sv_gridsearch_fit.best_estimator_.predict(X_test)

print(classification_report(y_test, y_preds_clf_sv_gridsearch))
```

### Fully-Connected Neural Network

Lastly, we look at the fully-connected neural network. Scikit-learn has an easy-to-use neural network module. The most important part of the classifier is setting the hidden layer sizes. Below, we use three hidden layers with the first layer being 450 hidden units, the second layer being 600 hidden units, and the third layer being 450 hidden units. These numbers were chosen arbitrarily; play around and see if you can do better.

```
# https://scikit-learn.org/stable/modules/neural_networks_supervised.html
from sklearn.neural_network import MLPClassifier

nn_clf = MLPClassifier(solver='adam', hidden_layer_sizes=(450,600,450))
nn_clf.fit(X_train, y_train)
```

```
y_preds_clf_nn = nn_clf.predict(X_test)

print(classification_report(y_test, y_preds_clf_nn))
```

### Standardization of Word Embedding Features

```
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()

X_train_scaled = std_scaler.fit_transform(X_train)
```

```
X_test_scaled = std_scaler.transform(X_test)

nn_clf = MLPClassifier(solver='adam', hidden_layer_sizes=(450,600,450))
nn_clf.fit(X_train_scaled, y_train)

y_preds_clf_nn = nn_clf.predict(X_test_scaled)

print(classification_report(y_test, y_preds_clf_nn))
```
